```python
M = 1000

signal_min_low_df = {}
signal_vals_low_df = {}
nucs_idx_low_df = {}
N_nucs_low_df = {}
peak_nuc_counts_low_df = {}

signal_min_mid_df = {}
signal_vals_mid_df = {}
nucs_idx_mid_df = {}
N_nucs_mid_df = {}
peak_nuc_counts_mid_df = {}

signal_min_high_df = {}
signal_vals_high_df = {}
nucs_idx_high_df = {}
N_nucs_high_df = {}
peak_nuc_counts_high_df = {}


for k in range(8):

    signal_min_low = np.zeros(M)
    signal_vals_low = []
    nucs_idx_low = []
    N_nucs_low = np.zeros(M)

    signal_min_mid = np.zeros(M)
    signal_vals_mid = []
    nucs_idx_mid = []
    N_nucs_mid = np.zeros(M)


    signal_min_high = np.zeros(M)
    signal_vals_high = []
    nucs_idx_high = []
    N_nucs_high = np.zeros(M)

    peak_nuc_counts_low = np.zeros(len(peaks_df[k]))
    peak_nuc_counts_mid = np.zeros(len(peaks_df[k]))
    peak_nuc_counts_high = np.zeros(len(peaks_df[k]))


    for m in range(M):

        nucs = []

        break_con = False

        for i in range(len(peaks_df[k])):
            rand_val = np.random.uniform()
            if i != 0:
                j = 0
                while (np.abs(location_df[k][(np.abs(cdf_df[k]-rand_val)).argmin()]-nucs).min() <= 2*half_wrap+1) and (not break_con):
                    rand_val = np.random.uniform()
                    j += 1
                    if j > 1000:
                        break_con = True
            if (not break_con):
                nucs = np.append(nucs,(np.abs(cdf_df[k]-rand_val)).argmin())
            else:
                break

        nucs_low = nucs[0:int(np.min([len(nucs),len(peaks_df[k])/4]))].astype(int)
        nucs_mid = nucs[0:int(np.min([len(nucs),len(peaks_df[k])/2]))].astype(int)
        nucs_high = nucs.astype(int)

        nucs_idx_low = np.append(nucs_idx_low,nucs_low)
        signal_vals_low = np.append(signal_vals_low,signal_df[k][nucs_low])
        signal_min_low[m] = np.min(signal_df[k][nucs_low])
        N_nucs_low[m] = len(nucs[0:int(np.min([len(nucs),len(peaks_df[k])/4]))])

        nucs_idx_mid = np.append(nucs_idx_mid,nucs_mid)
        signal_vals_mid = np.append(signal_vals_mid,signal_df[k][nucs_mid])
        signal_min_mid[m] = np.min(signal_df[k][nucs_mid])
        N_nucs_mid[m] = len(nucs[0:int(np.min([len(nucs),len(peaks_df[k])/2]))])

        nucs_idx_high = np.append(nucs_idx_high,nucs_high)
        signal_vals_high = np.append(signal_vals_high,signal_df[k][nucs_high])
        signal_min_high[m] = np.min(signal_df[k][nucs_high])
        N_nucs_high[m] = len(nucs)


        temp = np.zeros(len(peaks_df[k]))
        for j in range(len(nucs_low)):
            jx = np.argmin(np.abs(location_df[k][peaks_df[k]]-location_df[k][nucs_low[j].astype(int)]))
            if temp[jx] == 0:
                peak_nuc_counts_low[jx] += 1/M
                temp[jx] = 1
        temp = np.zeros(len(peaks_df[k]))
        for j in range(len(nucs_mid)):
            jx = np.argmin(np.abs(location_df[k][peaks_df[k]]-location_df[k][nucs_mid[j].astype(int)]))
            if temp[jx] == 0:
                peak_nuc_counts_mid[jx] += 1/M
                temp[jx] = 1
        temp = np.zeros(len(peaks_df[k]))
        for j in range(len(nucs_high)):
            jx = np.argmin(np.abs(location_df[k][peaks_df[k]]-location_df[k][nucs_high[j].astype(int)]))
            if temp[jx] == 0:
                peak_nuc_counts_high[jx] += 1/M
                temp[jx] = 1
                
    signal_min_low_df[k] = signal_min_low
    signal_vals_low_df[k] = signal_vals_low
    nucs_idx_low_df[k] = nucs_idx_low
    N_nucs_low_df[k] = N_nucs_low
    peak_nuc_counts_low_df[k] = peak_nuc_counts_low

    signal_min_mid_df[k] = signal_min_mid
    signal_vals_mid_df[k] = signal_vals_mid
    nucs_idx_mid_df[k] =  nucs_idx_mid
    N_nucs_mid_df[k] = N_nucs_mid
    peak_nuc_counts_mid_df[k] = peak_nuc_counts_mid

    signal_min_high_df[k] = signal_min_high
    signal_vals_high_df[k] = signal_vals_high
    nucs_idx_high_df[k] = nucs_idx_high
    N_nucs_high_df[k] = N_nucs_high
    peak_nuc_counts_high_df[k] = peak_nuc_counts_high
```
